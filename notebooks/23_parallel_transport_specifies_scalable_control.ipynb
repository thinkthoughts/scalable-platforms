## 1. Setup

This notebook is standalone and works in Colab, Jupyter, or a local repo clone. It writes PNGs to:

```text
figures/
results/23_parallel_transport_hardware/figures/
```


````python
from pathlib import Path
import json
import zipfile

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from matplotlib.patches import Circle, Rectangle, FancyArrowPatch

NOTEBOOK_NAME = "23_parallel_transport_hardware"
CWD = Path.cwd().resolve()

if (CWD / "notebooks").exists() or (CWD / "figures").exists():
    REPO_ROOT = CWD
elif CWD.name == "notebooks":
    REPO_ROOT = CWD.parent
else:
    REPO_ROOT = CWD

NOTEBOOKS = REPO_ROOT / "notebooks"
FIGURES = REPO_ROOT / "figures"
RESULTS = REPO_ROOT / "results" / NOTEBOOK_NAME
RESULT_FIGURES = RESULTS / "figures"

for path in [NOTEBOOKS, FIGURES, RESULTS, RESULT_FIGURES]:
    path.mkdir(parents=True, exist_ok=True)

def save_png(fig, name, dpi=220):
    repo_path = FIGURES / f"{name}.png"
    result_path = RESULT_FIGURES / f"{name}.png"
    fig.savefig(repo_path, dpi=dpi, bbox_inches="tight")
    fig.savefig(result_path, dpi=dpi, bbox_inches="tight")
    print(f"✓ Saved {repo_path}")
    print(f"✓ Saved {result_path}")
    return repo_path

def atom(ax, x, y, r=0.018, lw=1.6, fill=False):
    c = Circle((x, y), r, fill=fill, linewidth=lw, edgecolor="black", facecolor="white")
    ax.add_patch(c)
    return c

def rail(ax, x0, x1, y, lw=2.1):
    ax.plot([x0, x1], [y, y], linewidth=lw, solid_capstyle="round")

def arrow(ax, x0, y0, x1, y1, lw=1.5, mutation_scale=12):
    ax.add_patch(FancyArrowPatch((x0, y0), (x1, y1), arrowstyle="->",
                                 linewidth=lw, mutation_scale=mutation_scale, color="black"))

def label_box(ax, x, y, text, fs=11, weight="normal", pad=0.32):
    ax.text(x, y, text, ha="center", va="center", fontsize=fs, weight=weight,
            bbox=dict(boxstyle=f"round,pad={pad}", linewidth=1.2, facecolor="white"))

print(f"Repo root: {REPO_ROOT}")
print(f"Figures:   {FIGURES}")
print(f"Results:   {RESULTS}")
````

## 2. Variables and abstractions

Notebook 23 treats the transport hardware as the architecture-defining object.

\[
T_{\mathrm{coherent}}
\rightarrow
A_{\mathrm{scalable}}
\]

\[
B_{\mathrm{control}}
\sim
N_{\mathrm{rails}}
F_{\mathrm{transport}}
R_{\mathrm{schedule}}
\]

where \(T_{\mathrm{coherent}}\) is coherent transport, \(A_{\mathrm{scalable}}\) is the scalable architecture, \(N_{\mathrm{rails}}\) is the number of usable transport rails, \(F_{\mathrm{transport}}\) is transport fidelity, and \(R_{\mathrm{schedule}}\) is the schedulable routing rate.


````python
variables = pd.DataFrame([
    {"symbol": "T_coherent", "meaning": "coherent transport", "role": "state-preserving physical motion"},
    {"symbol": "A_scalable", "meaning": "scalable architecture", "role": "processor organization built from transport"},
    {"symbol": "N_rails", "meaning": "usable transport rails", "role": "parallel physical paths"},
    {"symbol": "F_transport", "meaning": "transport fidelity", "role": "preservation of stored state information"},
    {"symbol": "R_schedule", "meaning": "schedulable routing rate", "role": "conflict-free motion throughput"},
    {"symbol": "B_control", "meaning": "control bandwidth", "role": "useful simultaneous transport"},
])
variables_path = RESULTS / "parallel_transport_hardware_variables.csv"
variables.to_csv(variables_path, index=False)
variables
````

## 8. Transport scheduling on rails

Scheduling converts transport rails into controllable architecture bandwidth.


````python
schedule = pd.DataFrame([
    {"time": "t0", "rail": "rail 1", "task": "storage → gate"},
    {"time": "t0", "rail": "rail 2", "task": "storage → buffer"},
    {"time": "t0", "rail": "rail 3", "task": "idle"},
    {"time": "t1", "rail": "rail 1", "task": "gate → measurement"},
    {"time": "t1", "rail": "rail 2", "task": "buffer → gate"},
    {"time": "t1", "rail": "rail 3", "task": "storage → buffer"},
    {"time": "t2", "rail": "rail 1", "task": "reset"},
    {"time": "t2", "rail": "rail 2", "task": "gate → measurement"},
    {"time": "t2", "rail": "rail 3", "task": "buffer → gate"},
])
schedule_path = RESULTS / "transport_rail_schedule.csv"
schedule.to_csv(schedule_path, index=False)

fig, ax = plt.subplots(figsize=(13.5, 7.4))
ax.axis("off")

ax.text(0.5, 0.94, "Transport scheduling on rails", ha="center", va="center", fontsize=22, weight="bold")
ax.text(0.5, 0.875, "Independent rail motions become useful only when their timing avoids conflict.",
        ha="center", va="center", fontsize=13.5)

rail_ys = [0.66, 0.52, 0.38]
rail_labels = ["rail 1", "rail 2", "rail 3"]
for y, label in zip(rail_ys, rail_labels):
    ax.text(0.09, y, label, ha="right", va="center", fontsize=12.5, weight="bold")
    rail(ax, 0.16, 0.88, y, lw=1.8)

times = {"t0":0.28, "t1":0.52, "t2":0.76}
ax.text(0.28, 0.76, "t0", ha="center", va="center", fontsize=13, weight="bold")
ax.text(0.52, 0.76, "t1", ha="center", va="center", fontsize=13, weight="bold")
ax.text(0.76, 0.76, "t2", ha="center", va="center", fontsize=13, weight="bold")

tasks = {
    ("t0","rail 1"): "S→G",
    ("t0","rail 2"): "S→B",
    ("t0","rail 3"): "idle",
    ("t1","rail 1"): "G→M",
    ("t1","rail 2"): "B→G",
    ("t1","rail 3"): "S→B",
    ("t2","rail 1"): "reset",
    ("t2","rail 2"): "G→M",
    ("t2","rail 3"): "B→G",
}

for t, x in times.items():
    for y, label in zip(rail_ys, rail_labels):
        txt = tasks[(t,label)]
        if txt != "idle" and txt != "reset":
            atom(ax, x-0.035, y, r=0.011, fill=True)
            arrow(ax, x-0.018, y, x+0.055, y, lw=1.2, mutation_scale=10)
        label_box(ax, x, y+0.045, txt, fs=10.5, pad=0.22)

arrow(ax, 0.34, 0.20, 0.66, 0.20, lw=1.5)
ax.text(0.50, 0.16, "time", ha="center", va="center", fontsize=11.5)

ax.text(0.50, 0.08,
        "A scheduler turns parallel physical rails into simultaneous usable operations.",
        ha="center", va="center", fontsize=12.5)

ax.set_xlim(0,1)
ax.set_ylim(0,1)
fig.tight_layout()

schedule_fig = save_png(fig, "23_transport_schedule_rails")
plt.show()
schedule
````

## 9. Architecture scaling

Adding rails and zones grows the controllable processor.


````python
scaling = pd.DataFrame([
    {"case": "small", "storage_sites": 4, "transport_rails": 1, "gate_sites": 2},
    {"case": "medium", "storage_sites": 8, "transport_rails": 2, "gate_sites": 4},
    {"case": "large", "storage_sites": 16, "transport_rails": 4, "gate_sites": 8},
])
scaling_path = RESULTS / "architecture_scaling.csv"
scaling.to_csv(scaling_path, index=False)

fig, ax = plt.subplots(figsize=(13.5, 7.6))
ax.axis("off")

ax.text(0.5, 0.94, "Architecture scaling through transport hardware",
        ha="center", va="center", fontsize=22, weight="bold")
ax.text(0.5, 0.875, "Processors scale by adding storage sites, transport rails, and interaction zones.",
        ha="center", va="center", fontsize=13.5)

cases = [
    ("small", 0.72, 4, 1, 2),
    ("medium", 0.50, 8, 2, 4),
    ("large", 0.25, 12, 4, 6),
]

for label, y, n_storage, n_rails, n_gates in cases:
    ax.text(0.08, y, label, ha="right", va="center", fontsize=14, weight="bold")
    # storage atoms
    xs = np.linspace(0.18, 0.46, n_storage)
    for x in xs:
        atom(ax, x, y+0.055, r=0.009)
    # rails
    if n_rails == 1:
        rail_ys = [y]
    else:
        rail_ys = np.linspace(y-0.035, y+0.035, n_rails)
    for yy in rail_ys:
        rail(ax, 0.18, 0.46, yy, lw=1.5)
    # gates
    gate_xs = np.linspace(0.62, 0.84, n_gates)
    for x in gate_xs:
        atom(ax, x, y, r=0.009)
    arrow(ax, 0.49, y, 0.59, y, lw=1.2, mutation_scale=10)

ax.text(0.32, 0.82, "storage", ha="center", va="center", fontsize=12, weight="bold")
ax.text(0.32, 0.13, "transport rails", ha="center", va="center", fontsize=12, weight="bold")
ax.text(0.73, 0.82, "gate zones", ha="center", va="center", fontsize=12, weight="bold")

ax.text(0.50, 0.06,
        "Scalable architecture means more useful regions connected by coherent, schedulable motion.",
        ha="center", va="center", fontsize=12.5)

ax.set_xlim(0,1)
ax.set_ylim(0,1)
fig.tight_layout()

scaling_fig = save_png(fig, "23_architecture_scaling")
plt.show()
scaling
````

Run the final cell to package the notebook outputs for download.


````python
from IPython.display import FileLink, display

zip_path = RESULTS / "23_parallel_transport_hardware_artifacts.zip"

paths_to_include = [
    FIGURES,
    RESULTS,
]

with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as zf:
    for item in paths_to_include:
        item = Path(item)
        if item.is_dir():
            for path in item.rglob("*"):
                if not path.is_file():
                    continue
                if path.resolve() == zip_path.resolve():
                    continue
                try:
                    archive_name = path.relative_to(REPO_ROOT)
                except ValueError:
                    archive_name = path.name
                zf.write(path, archive_name)
        elif item.exists():
            try:
                archive_name = item.relative_to(REPO_ROOT)
            except ValueError:
                archive_name = item.name
            zf.write(item, archive_name)

print(f"Created: {zip_path}")
print(f"Size: {zip_path.stat().st_size:,} bytes")
display(FileLink(str(zip_path)))

try:
    from google.colab import files
    files.download(str(zip_path))
except Exception:
    pass
````
