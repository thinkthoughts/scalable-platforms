## 1. Setup

````python
from pathlib import Path
import json
import zipfile

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

NOTEBOOK_NAME = "17_coherent_transport"
CWD = Path.cwd().resolve()

if (CWD / "notebooks").exists() or (CWD / "figures").exists():
    REPO_ROOT = CWD
elif CWD.name == "notebooks":
    REPO_ROOT = CWD.parent
else:
    REPO_ROOT = CWD

NOTEBOOKS = REPO_ROOT / "notebooks"
FIGURES = REPO_ROOT / "figures"
RESULTS = REPO_ROOT / "results" / NOTEBOOK_NAME
RESULT_FIGURES = RESULTS / "figures"

for path in [NOTEBOOKS, FIGURES, RESULTS, RESULT_FIGURES]:
    path.mkdir(parents=True, exist_ok=True)

def save_png(fig, name, dpi=220):
    repo_path = FIGURES / f"{name}.png"
    result_path = RESULT_FIGURES / f"{name}.png"
    fig.savefig(repo_path, dpi=dpi, bbox_inches="tight")
    fig.savefig(result_path, dpi=dpi, bbox_inches="tight")
    print(f"✓ Saved {repo_path}")
    print(f"✓ Saved {result_path}")
    return repo_path

print(f"Repo root: {REPO_ROOT}")
print(f"Figures:   {FIGURES}")
print(f"Results:   {RESULTS}")
````

## 2. Variables and abstractions

Notebook 17 treats transport as a fidelity-preserving operation.

\[
F_{\mathrm{transport}}
\rightarrow
D_{\mathrm{control}}
\]

\[
F_{\mathrm{transport}}
=
1-\epsilon_{\mathrm{motion}}-\epsilon_{\mathrm{heating}}-\epsilon_{\mathrm{loss}}-\epsilon_{\mathrm{phase}}
\]

where \(F_{\mathrm{transport}}\) is transport fidelity, \(D_{\mathrm{control}}\) is reliable control depth, and the error terms describe motion error, heating, loss, and phase error.


````python
variables = pd.DataFrame([
    {"symbol": "F_transport", "meaning": "transport fidelity", "role": "state preservation during motion"},
    {"symbol": "D_control", "meaning": "reliable control depth", "role": "useful control after motion"},
    {"symbol": "epsilon_motion", "meaning": "motional excitation error", "role": "transport cost"},
    {"symbol": "epsilon_heating", "meaning": "heating during motion", "role": "coherence cost"},
    {"symbol": "epsilon_loss", "meaning": "atom loss during transport", "role": "layout cost"},
    {"symbol": "epsilon_phase", "meaning": "phase error accumulated in motion", "role": "state-memory cost"},
    {"symbol": "M_transport", "meaning": "transport fidelity margin", "role": "fault-tolerance headroom"},
])
variables_path = RESULTS / "transport_variables.csv"
variables.to_csv(variables_path, index=False)
variables
````

## 4. Transport pipeline

Transport is not merely a mechanical operation. It is part of the computation stack because every movement must preserve the stored quantum information.


````python
pipeline = [
    ("Hyperfine\nstate", "stored information"),
    ("Store\natom", "prepare memory"),
    ("Move\natom", "transport coherently"),
    ("Address\natom", "select location"),
    ("Gate\noperation", "compute"),
    ("Measurement", "read out"),
]

fig, ax = plt.subplots(figsize=(10.5, 7.2))
ax.axis("off")

ax.text(0.5, 0.95, "Transport pipeline", ha="center", va="center", fontsize=20, weight="bold")
ax.text(0.5, 0.88, "Every transport step must preserve the stored quantum information.", ha="center", va="center", fontsize=12.5)

x = 0.5
y0 = 0.76
dy = 0.125

for i, (title, subtitle) in enumerate(pipeline):
    y = y0 - i * dy
    ax.text(
        x, y,
        f"{title}\n{subtitle}",
        ha="center", va="center",
        fontsize=12.5,
        bbox=dict(boxstyle="round,pad=0.55", linewidth=1.2, facecolor="white"),
    )
    if i < len(pipeline) - 1:
        ax.annotate("", xy=(x, y - 0.075), xytext=(x, y - 0.045),
                    arrowprops=dict(arrowstyle="->", linewidth=1.4))

ax.text(0.5, 0.08, "state memory  →  motion  →  addressing  →  gates  →  measurement", ha="center", va="center", fontsize=13)

fig.tight_layout()
pipeline_fig = save_png(fig, "17_transport_pipeline")
plt.show()
pipeline_fig
````

````python
fig, ax = plt.subplots(figsize=(12, 6.4))
ax.axis("off")

ax.text(0.5, 0.94, "Moving atoms through the processor", ha="center", va="center", fontsize=20, weight="bold")
ax.text(0.5, 0.87, "Transport connects storage, gate, and measurement regions.", ha="center", va="center", fontsize=12.5)

# zones
zones = [
    (0.18, 0.62, 0.22, 0.18, "Storage zone"),
    (0.50, 0.62, 0.24, 0.18, "Transport channel"),
    (0.80, 0.62, 0.18, 0.18, "Gate zone"),
    (0.80, 0.28, 0.18, 0.16, "Measurement"),
]
for x, y, w, h, label in zones:
    rect = plt.Rectangle((x - w/2, y - h/2), w, h, fill=False, linewidth=1.4)
    ax.add_patch(rect)
    ax.text(x, y + h/2 + 0.035, label, ha="center", va="center", fontsize=12, weight="bold")

# atoms
storage_atoms = [(0.12,0.61),(0.17,0.65),(0.22,0.61),(0.17,0.57)]
for x,y in storage_atoms:
    circ = plt.Circle((x,y), 0.016, fill=False, linewidth=1.3)
    ax.add_patch(circ)

# channel
ax.plot([0.38, 0.62], [0.62, 0.62], linewidth=2.0)
for x in [0.42, 0.50, 0.58]:
    ax.text(x, 0.66, "•", ha="center", va="center", fontsize=18)

# gate pair
for x,y in [(0.77,0.62),(0.83,0.62)]:
    circ = plt.Circle((x,y), 0.018, fill=False, linewidth=1.5)
    ax.add_patch(circ)
ax.plot([0.77,0.83],[0.62,0.62], linewidth=1.2)

# arrows
ax.annotate("", xy=(0.37,0.62), xytext=(0.29,0.62), arrowprops=dict(arrowstyle="->", linewidth=1.5))
ax.annotate("", xy=(0.70,0.62), xytext=(0.62,0.62), arrowprops=dict(arrowstyle="->", linewidth=1.5))
ax.annotate("", xy=(0.80,0.39), xytext=(0.80,0.52), arrowprops=dict(arrowstyle="->", linewidth=1.5))

ax.text(0.5, 0.10, "Transport turns physical separation into a scalable control architecture.", ha="center", va="center", fontsize=13)

ax.set_xlim(0,1)
ax.set_ylim(0,1)
fig.tight_layout()

layout_fig = save_png(fig, "17_processor_layout")
plt.show()
layout_fig
````

## 6. Transport error budget

Notebook 13 used a coherence budget. Notebook 17 uses an error budget: transport remains useful when motion, heating, loss, and phase error leave enough fidelity margin.


````python
budget_scenarios = pd.DataFrame([
    {"scenario": "large margin", "motion": 0.10, "heating": 0.05, "loss": 0.04, "phase": 0.06, "margin": 0.75},
    {"scenario": "medium margin", "motion": 0.18, "heating": 0.10, "loss": 0.08, "phase": 0.14, "margin": 0.50},
    {"scenario": "small margin", "motion": 0.28, "heating": 0.16, "loss": 0.14, "phase": 0.22, "margin": 0.20},
])
budget_scenarios_path = RESULTS / "transport_error_budget.csv"
budget_scenarios.to_csv(budget_scenarios_path, index=False)

fig, ax = plt.subplots(figsize=(12.5, 6.6))
ax.axis("off")

ax.text(0.5, 0.94, "Transport error budget", ha="center", va="center", fontsize=20, weight="bold")
ax.text(0.5, 0.87, "The remaining fidelity margin is the engineering resource.", ha="center", va="center", fontsize=13)

categories = ["motion", "heating", "loss", "phase", "margin"]
labels = {
    "motion": "motion",
    "heating": "heating",
    "loss": "loss",
    "phase": "phase",
    "margin": "remaining margin",
}

left = 0.16
total_width = 0.70
bar_h = 0.075
ys = [0.68, 0.48, 0.28]

for row_idx, row in budget_scenarios.iterrows():
    y = ys[row_idx]
    ax.text(0.09, y + bar_h/2, row["scenario"], ha="right", va="center", fontsize=12, weight="bold")
    x = left
    for cat in categories:
        w = total_width * row[cat]
        face = "white" if cat != "margin" else "#F7F8FA"
        rect = plt.Rectangle((x, y), w, bar_h, fill=True, facecolor=face, edgecolor="black", linewidth=1.2)
        ax.add_patch(rect)
        if w > 0.055:
            ax.text(x + w/2, y + bar_h/2, labels[cat], ha="center", va="center",
                    fontsize=9 if cat != "margin" else 10, weight="bold" if cat == "margin" else "normal")
        x += w

    margin_start = left + total_width * (1 - row["margin"])
    margin_mid = margin_start + total_width * row["margin"] / 2
    ax.annotate("", xy=(margin_mid, y - 0.025), xytext=(margin_mid, y - 0.07),
                arrowprops=dict(arrowstyle="->", linewidth=1.1))
    ax.text(margin_mid, y - 0.095, f"{row['margin']:.0%} margin", ha="center", va="center", fontsize=10)

ax.text(0.5, 0.08, "Better transport preserves fidelity margin for addressing, gates, and logical computation.", ha="center", va="center", fontsize=12)

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
fig.tight_layout()

budget_fig = save_png(fig, "17_transport_error_budget")
plt.show()
budget_scenarios
````

## 7. Transport fidelity

Transport fidelity depends on distance, speed, noise, and control optimization. This abstraction compares three transport regimes.


````python
distance = np.linspace(0, 1.0, 101)
transport_curves = pd.DataFrame({
    "distance": distance,
    "optimized": 1 - 0.0004 * distance**1.2,
    "adiabatic": 1 - 0.0012 * distance**1.5,
    "noisy": 1 - 0.0040 * distance**1.1,
})
fidelity_path = RESULTS / "transport_fidelity.csv"
transport_curves.to_csv(fidelity_path, index=False)

fig, ax = plt.subplots(figsize=(9.4, 5.4))
ax.plot(transport_curves["distance"], transport_curves["optimized"], label="optimized")
ax.plot(transport_curves["distance"], transport_curves["adiabatic"], label="adiabatic")
ax.plot(transport_curves["distance"], transport_curves["noisy"], label="noisy")

ax.set_xlabel("Transport distance")
ax.set_ylabel("Transport fidelity")
ax.set_title("Transport engineering preserves fidelity over distance")
ax.set_ylim(0.995, 1.0002)
ax.legend(title="Regime")
fig.tight_layout()

fidelity_fig = save_png(fig, "17_transport_fidelity")
plt.show()
transport_curves.head()
````

## 13. Download artifacts

Run the final cell to package the notebook outputs for download.


````python
from IPython.display import FileLink, display

zip_path = RESULTS / "17_coherent_transport_artifacts.zip"

paths_to_include = [
    FIGURES,
    RESULTS,
]

with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as zf:
    for item in paths_to_include:
        item = Path(item)
        if item.is_dir():
            for path in item.rglob("*"):
                if not path.is_file():
                    continue
                if path.resolve() == zip_path.resolve():
                    continue
                try:
                    archive_name = path.relative_to(REPO_ROOT)
                except ValueError:
                    archive_name = path.name
                zf.write(path, archive_name)
        elif item.exists():
            try:
                archive_name = item.relative_to(REPO_ROOT)
            except ValueError:
                archive_name = item.name
            zf.write(item, archive_name)

print(f"Created: {zip_path}")
print(f"Size: {zip_path.stat().st_size:,} bytes")
display(FileLink(str(zip_path)))

try:
    from google.colab import files
    files.download(str(zip_path))
except Exception:
    pass
````
