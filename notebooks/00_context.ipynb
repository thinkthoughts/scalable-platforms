## 5. Engineering constraints

Coherence alone is not enough. Scaling introduces constraints that determine which platform paths remain practical.

Notebook 00 uses six shared constraints:

\[
\{\text{decoherence},\text{noise},\text{heating},\text{loss},\text{crosstalk},\text{stability}\}
\]

Later notebooks can attach measurements, models, or design rules to each one.


```python
constraints = pd.DataFrame([
    {"constraint": "decoherence", "computation": "state errors during storage, transport, and gates", "metrology": "loss of phase information during interrogation"},
    {"constraint": "noise", "computation": "laser, magnetic, motional, and technical fluctuations", "metrology": "line pulling, field sensitivity, readout noise"},
    {"constraint": "heating", "computation": "motional excitation during rearrangement or transport", "metrology": "trap-induced temperature and motional shifts"},
    {"constraint": "loss", "computation": "atom loss, imaging loss, finite filling", "metrology": "molecule loss, trap lifetime, detection loss"},
    {"constraint": "crosstalk", "computation": "unwanted addressing of neighboring sites", "metrology": "unwanted coupling between internal states"},
    {"constraint": "stability", "computation": "array, laser, and control stability across large systems", "metrology": "frequency, wavelength, and systematic stability"},
])
constraints_path = RESULTS / "engineering_constraints.csv"
constraints.to_csv(constraints_path, index=False)
constraints
```

```python
fig, ax = plt.subplots(figsize=(10.5, 6))
ax.axis("off")
ax.text(0.5, 0.94, "Engineering constraints set the useful operating regime", ha="center", va="center", fontsize=17, weight="bold")

# Top chain
chain = ["Coherence", "Control", "Scale", "Performance"]
for i, label in enumerate(chain):
    x = 0.18 + i * 0.21
    ax.text(x, 0.80, label, ha="center", va="center", fontsize=12,
            bbox=dict(boxstyle="round,pad=0.45", linewidth=1.2, facecolor="white"))
    if i < len(chain) - 1:
        ax.annotate("", xy=(x+0.13, 0.80), xytext=(x+0.08, 0.80), arrowprops=dict(arrowstyle="->", linewidth=1.4))

# Constraint grid
items = constraints["constraint"].tolist()
positions = [(0.25,0.58),(0.5,0.58),(0.75,0.58),(0.25,0.38),(0.5,0.38),(0.75,0.38)]
for (x,y), item in zip(positions, items):
    ax.text(x, y, item.upper(), ha="center", va="center", fontsize=12, weight="bold",
            bbox=dict(boxstyle="round,pad=0.65", linewidth=1.2, facecolor="#F7F8FA"))

ax.text(0.5, 0.16, "larger systems  →  longer coherence  →  better control  →  lower uncertainty", 
        ha="center", va="center", fontsize=12)
fig.tight_layout()

constraints_fig = FIGURES / "00_engineering_constraints.png"
fig.savefig(constraints_fig, dpi=220, bbox_inches="tight")
plt.show()
constraints_fig
```

## 11. Download artifacts

Run the final cell to package the notebook outputs for download.


```python
# Package notebook artifacts for download
from pathlib import Path
from IPython.display import FileLink, display
import zipfile

zip_path = RESULTS / "00_context_artifacts.zip"

notebook_path = NOTEBOOKS / "00_context.ipynb"
fallback_notebook_path = Path.cwd() / "00_context.ipynb"

paths_to_include = [
    notebook_path if notebook_path.exists() else fallback_notebook_path,
    FIGURES,
    RESULTS,
    SITE,
    SRC,
    REPO_ROOT / "README.md",
    REPO_ROOT / "requirements.txt",
]

with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as zf:
    for item in paths_to_include:
        item = Path(item)
        if item.is_dir():
            for path in item.rglob("*"):
                if not path.is_file():
                    continue
                if path.resolve() == zip_path.resolve():
                    continue
                try:
                    archive_name = path.relative_to(REPO_ROOT)
                except ValueError:
                    archive_name = path.name
                zf.write(path, archive_name)
        elif item.exists():
            if item.resolve() == zip_path.resolve():
                continue
            try:
                archive_name = item.relative_to(REPO_ROOT)
            except ValueError:
                archive_name = item.name
            zf.write(item, archive_name)

print(f"Created: {zip_path}")
print(f"Size: {zip_path.stat().st_size:,} bytes")
display(FileLink(str(zip_path)))

try:
    from google.colab import files
    files.download(str(zip_path))
except Exception:
    pass
```
