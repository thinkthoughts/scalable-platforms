## 5. Engineering constraints

Coherence alone leaves the engineering problem incomplete. Scaling introduces constraints that determine which platform paths remain practical.

Notebook 00 uses six shared constraints:

\[
\{\text{decoherence},\text{noise},\text{heating},\text{loss},\text{crosstalk},\text{stability}\}
\]

Later notebooks can attach measurements, models, or design rules to each one.


```python
constraints = pd.DataFrame([
    {"constraint": "decoherence", "computation": "state errors during storage, transport, and gates", "metrology": "loss of phase information during interrogation"},
    {"constraint": "noise", "computation": "laser, magnetic, motional, and technical fluctuations", "metrology": "line pulling, field sensitivity, readout noise"},
    {"constraint": "heating", "computation": "motional excitation during rearrangement or transport", "metrology": "trap-induced temperature and motional shifts"},
    {"constraint": "loss", "computation": "atom loss, imaging loss, finite filling", "metrology": "molecule loss, trap lifetime, detection loss"},
    {"constraint": "crosstalk", "computation": "unwanted addressing of neighboring sites", "metrology": "unwanted coupling between internal states"},
    {"constraint": "stability", "computation": "array, laser, and control stability across large systems", "metrology": "frequency, wavelength, and systematic stability"},
])

constraints_path = RESULTS / "engineering_constraints.csv"
constraints.to_csv(constraints_path, index=False)
constraints
```

```python
fig, ax = plt.subplots(figsize=(10.5, 6))
ax.axis("off")
ax.text(0.5, 0.94, "Engineering constraints set the useful operating regime", ha="center", va="center", fontsize=17, weight="bold")

chain = ["Coherence", "Control", "Scale", "Performance"]
for i, label in enumerate(chain):
    x = 0.18 + i * 0.21
    ax.text(
        x, 0.80, label,
        ha="center", va="center",
        fontsize=12,
        bbox=dict(boxstyle="round,pad=0.45", linewidth=1.2, facecolor="white"),
    )
    if i < len(chain) - 1:
        ax.annotate("", xy=(x + 0.13, 0.80), xytext=(x + 0.08, 0.80), arrowprops=dict(arrowstyle="->", linewidth=1.4))

items = constraints["constraint"].tolist()
positions = [(0.25, 0.58), (0.5, 0.58), (0.75, 0.58), (0.25, 0.38), (0.5, 0.38), (0.75, 0.38)]
for (x, y), item in zip(positions, items):
    ax.text(
        x, y, item.upper(),
        ha="center", va="center",
        fontsize=12,
        weight="bold",
        bbox=dict(boxstyle="round,pad=0.65", linewidth=1.2, facecolor="#F7F8FA"),
    )

ax.text(
    0.5, 0.16,
    "larger systems  →  longer coherence  →  better control  →  lower uncertainty",
    ha="center", va="center",
    fontsize=12,
)
fig.tight_layout()

saved = save_png(fig, "00_engineering_constraints", PATHS, dpi=220)
constraints_fig = saved["figures"]

plt.show()
constraints_fig
```

## 6. Platform performance

For this repository, "performance" is deliberately plural.

Computation and metrology optimize different measurements, but both require coherence to survive the scaling process.

| Branch | Representative performance quantities |
|---|---|
| Computation | coherence time, transport efficiency, gate fidelity, filling fraction |
| Metrology | clock uncertainty, systematic shifts, linewidth, trap lifetime |


```python
performance = pd.DataFrame([
    {"branch": "computation", "quantity": "coherence time", "role": "preserve state information while operations accumulate"},
    {"branch": "computation", "quantity": "transport efficiency", "role": "move atoms without losing usable state information"},
    {"branch": "computation", "quantity": "gate fidelity", "role": "support logical operations and error correction"},
    {"branch": "computation", "quantity": "filling fraction", "role": "scale arrays without excessive rearrangement overhead"},
    {"branch": "metrology", "quantity": "clock uncertainty", "role": "bound systematic and statistical error"},
    {"branch": "metrology", "quantity": "frequency shift", "role": "identify trap, field, and motional perturbations"},
    {"branch": "metrology", "quantity": "linewidth", "role": "resolve stable vibrational transitions"},
    {"branch": "metrology", "quantity": "trap lifetime", "role": "maintain molecules through interrogation"},
])

performance_path = RESULTS / "platform_performance_quantities.csv"
performance.to_csv(performance_path, index=False)
performance
```

```python
fig, ax = plt.subplots(figsize=(11.5, 5.5))
ax.axis("off")
ax.text(0.5, 0.93, "Measured performance translates coherence into engineering progress", ha="center", fontsize=16, weight="bold")

columns = {
    "Computation": ["coherence time", "transport efficiency", "gate fidelity", "filling fraction"],
    "Metrology": ["clock uncertainty", "frequency shifts", "linewidth", "trap lifetime"],
}
xs = [0.30, 0.70]

for x, (title, items) in zip(xs, columns.items()):
    ax.text(
        x, 0.78, title.upper(),
        ha="center", va="center",
        fontsize=13,
        weight="bold",
        bbox=dict(boxstyle="round,pad=0.5", linewidth=1.2, facecolor="white"),
    )
    for j, item in enumerate(items):
        ax.text(
            x, 0.62 - j * 0.115,
            item,
            ha="center", va="center",
            fontsize=11,
            bbox=dict(boxstyle="round,pad=0.35", linewidth=0.9, facecolor="#F7F8FA"),
        )

ax.text(
    0.5, 0.08,
    "Sustainable collaboration preserves coherence while scaling systems.",
    ha="center", va="center",
    fontsize=13,
)
fig.tight_layout()

saved = save_png(fig, "00_measured_performance", PATHS, dpi=220)
performance_fig = saved["figures"]

plt.show()
performance_fig
```
