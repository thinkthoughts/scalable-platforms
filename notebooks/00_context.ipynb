## 1. Start here

This repository studies one engineering question:

> **How does coherence become a resource for scaling controllable platforms?**

\[

\text{next steps}

> **Coherence shifts from measured property to engineering resource.**


```python
from pathlib import Path
import json
import zipfile

import matplotlib.pyplot as plt
import pandas as pd

NOTEBOOK_NAME = "00_context"
CWD = Path.cwd().resolve()

# Standalone Colab/Jupyter-safe paths.
if (CWD / "notebooks").exists() or (CWD / "figures").exists():
    REPO_ROOT = CWD
elif CWD.name == "notebooks":
    REPO_ROOT = CWD.parent
else:
    REPO_ROOT = CWD

NOTEBOOKS = REPO_ROOT / "notebooks"
FIGURES = REPO_ROOT / "figures"
RESULTS = REPO_ROOT / "results" / NOTEBOOK_NAME
RESULT_FIGURES = RESULTS / "figures"

for path in [NOTEBOOKS, FIGURES, RESULTS, RESULT_FIGURES]:
    path.mkdir(parents=True, exist_ok=True)

def save_png(fig, name, dpi=220):
    repo_path = FIGURES / f"{name}.png"
    result_path = RESULT_FIGURES / f"{name}.png"
    fig.savefig(repo_path, dpi=dpi, bbox_inches="tight")
    fig.savefig(result_path, dpi=dpi, bbox_inches="tight")
    print(f"✓ Saved {repo_path}")
    print(f"✓ Saved {result_path}")
    return repo_path

print(f"Repo root: {REPO_ROOT}")
print(f"Figures:   {FIGURES}")
print(f"Results:   {RESULTS}")
```

## 8. Key equations

Platform performance as a constrained engineering outcome:

\[
P_{\mathrm{platform}}
=
f\left(
  C_{\phi},
  U,
  T,
  L,
  N,
  X,
  S,
  M
\right)
\]

where \(C_{\phi}\) denotes coherence, \(U\) control operations, \(T\) transport or trapping, \(L\) loss, \(N\) noise, \(X\) crosstalk, \(S\) stability, and \(M\) measurement.

Leading specification to measured outcomes:

\[
C_{\phi}
\longrightarrow
C_{\mathrm{eng}}
\longrightarrow
R_{\mathrm{platform}}
\longrightarrow
\{\tau, F, \eta_{\mathrm{transport}}, \delta\nu, \sigma, \ell\}

```python
from IPython.display import FileLink, display

zip_path = RESULTS / "00_context_artifacts.zip"

with zipfile.ZipFile(zip_path, "w", compression=zipfile.ZIP_DEFLATED) as zf:
    for item in [FIGURES, RESULTS]:
        item = Path(item)
        if item.is_dir():
            for path in item.rglob("*"):
                if path.is_file() and path.resolve() != zip_path.resolve():
                    try:
                        archive_name = path.relative_to(REPO_ROOT)
                    except ValueError:
                        archive_name = path.name
                    zf.write(path, archive_name)

print(f"Created: {zip_path}")
print(f"Size: {zip_path.stat().st_size:,} bytes")
display(FileLink(str(zip_path)))

try:
    from google.colab import files
    files.download(str(zip_path))
except Exception:
    pass
```
